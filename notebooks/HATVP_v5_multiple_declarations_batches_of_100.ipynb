```python
import requests
from lxml import etree
from tqdm import tqdm

def write_elements_to_file(elements, file_name):
    root = etree.Element("declarations")
    for element in elements:
        root.append(element)
    tree = etree.ElementTree(root)
    tree.write(file_name, pretty_print=True)

def batch_elements(elements, batch_size):
    for i in range(0, len(elements), batch_size):
        yield elements[i:i+batch_size]

def download_xml_file(url):
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get("content-length", 0))
    block_size = 1024
    progress_bar = tqdm(total=total_size, unit="B", unit_scale=True, desc="Downloading")

    with open("declarations.xml", "wb") as file:
        for data in response.iter_content(block_size):
            progress_bar.update(len(data))
            file.write(data)

    progress_bar.close()

# def process_declarations(declaration_number, output_path):

# # Parse the downloaded XML file
#     tree = etree.parse("declarations.xml")
#     root = tree.getroot()

#     # Select all elements inside "applications" tag
#     applications = root.findall("declaration")
#     print(applications)

#     batch_size = declaration_number
#     total_batches = len(applications) // batch_size

#     for idx, batch in tqdm(enumerate(batch_elements(applications, batch_size)), total=total_batches, desc="Processing batches"):
#         write_elements_to_file(batch, f"{output_path}declarations_hatvp_batch_{idx+1}.xml")


def process_declarations(declaration_number, output_path):
    # Parse the downloaded XML file
    tree = etree.parse("/content/declarations.xml")
    root = tree.getroot()

    # Select all elements inside "applications" tag
    applications = root.findall("declaration")
    print(applications)

    batch_size = declaration_number
    total_batches = len(applications) // batch_size

    for idx, batch in tqdm(enumerate(batch_elements(applications, batch_size)), total=total_batches, desc="Processing batches"):
        formatted_idx = str(idx + 1).zfill(8)  # pads with zeros
        write_elements_to_file(batch, f"{output_path}declarations_hatvp_batch_{formatted_idx}.xml")

output_path = "/content/xml_unitary_declarations/"
process_declarations(1, output_path)
```
