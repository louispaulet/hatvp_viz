```python
from lxml import etree

def write_elements_to_file(elements, file_name):
    root = etree.Element("declarations")
    for element in elements:
        root.append(element)
    tree = etree.ElementTree(root)
    tree.write(file_name, pretty_print=True)

def batch_elements(elements, batch_size):
    for i in range(0, len(elements), batch_size):
        yield elements[i:i+batch_size]

def main():
    # Parse the XML file
    tree = etree.parse("/content/drive/MyDrive/AI/HATVP/datasets/declarations.xml")
    root = tree.getroot()

    # Select all elements inside "applications" tag
    applications = root.findall("declaration")
    print(applications)

    # Save elements by batches of 100 to a dedicated xml file
    for idx, batch in enumerate(batch_elements(applications, 100)):
        write_elements_to_file(batch, f"/content/drive/MyDrive/AI/HATVP/datasets/xml_batches/declarations_hatvp_batch_{idx+1}.xml")
    
if __name__ == "__main__":
    main()
```
