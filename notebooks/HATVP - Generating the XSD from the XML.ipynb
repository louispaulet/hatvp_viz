```python
import pandas as pd
import xml.etree.ElementTree as ET

def get_xpaths(node, path='', xpaths=[]):
    current_path = f"{path}/{node.tag}"
    if len(node) == 0:
        xpaths.append(current_path)
    else:
        for child in node:
            get_xpaths(child, current_path, xpaths)
    
    return xpaths

# Load XML file
xml_file_path = '/content/drive/MyDrive/AI/HATVP/datasets/declarations.xml'
tree = ET.parse(xml_file_path)
root = tree.getroot()

# Get all unique xpaths
xpaths = get_xpaths(root)

# Create DataFrame
df = pd.DataFrame({'xpath': xpaths})

# Display the DataFrame
print(df)
```

```python
import pandas as pd
import xml.etree.ElementTree as ET

def get_xpaths(node, path='', xpaths={}):
    current_path = f"{path}/{node.tag}"
    if len(node) == 0:
        xpaths[current_path] = xpaths.get(current_path, 0) + 1
    else:
        for child in node:
            get_xpaths(child, current_path, xpaths)
    
    return xpaths

# Load XML file
xml_file_path = '/content/drive/MyDrive/AI/HATVP/datasets/declarations.xml'
tree = ET.parse(xml_file_path)
root = tree.getroot()

# Get all unique xpaths with occurrence counts
xpaths = get_xpaths(root)

# Create DataFrame
df = pd.DataFrame({'xpath': list(xpaths.keys()), 'occurrences': list(xpaths.values())})

# Display the DataFrame
df
```

```python
import pandas as pd
import xml.etree.ElementTree as ET

def get_declaration_values(node, declaration_values, current_values):
    for child in node:
        if len(child) == 0:
            xpath = f"{child.tag}"
            value = child.text
            current_values[xpath] = value
        else:
            get_declaration_values(child, declaration_values, current_values.copy())
    if len(current_values) > 0:
        declaration_values.append(current_values)

# Load XML file
xml_file_path = '/content/drive/MyDrive/AI/HATVP/datasets/declarations.xml'
tree = ET.parse(xml_file_path)
root = tree.getroot()

# Initialize list to store declaration values
declaration_values = []

# Extract values for each declaration object
for declaration in root.findall('.//declaration'):
    current_values = {}
    get_declaration_values(declaration, declaration_values, current_values)

# Create DataFrame
df_values = pd.DataFrame(declaration_values)

# Display the DataFrame
print(df_values)
```
