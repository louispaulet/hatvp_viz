```python
import xml.etree.ElementTree as ET
import pandas as pd

def extract_xml_location_and_id_to_df(xml_data, uuid = 42):
    # Parse the XML data
    root = ET.fromstring(xml_data)

    # Prepare dictionary to hold data
    data = {}

    # Specify the sections to parse
    sections_to_parse = ['.//general/qualiteMandat', './/general/organe', './/general/declarant']

    # Iterate through child elements and record text
    for section in sections_to_parse:
        for child in root.findall(section):
            for element in child:
                # Check if tag is already recorded. If not, initialize as empty list.
                if element.tag not in data:
                    data[element.tag] = []
                # Append text content if available. Otherwise append None.
                if element.text is not None:
                    data[element.tag].append(element.text.strip())
                    data[element.tag] = [' '.join(data[element.tag])]
                else:
                    data[element.tag].append(None)

    # Create a DataFrame
    df = pd.DataFrame(data)
    df.insert(0, 'uuid', uuid)
    # Return the DataFrame
    return df
```

## Extract mandat electif revenue (`mandatElectifDto`)

```python
def parse_mandatElectifDto(xml_data, uuid=42):
    # Parse the XML data
    root = ET.fromstring(xml_data)

    # Prepare list to hold data
    data = []

    # Find all 'items' in 'mandatElectifDto' section
    mandates = root.findall('.//mandatElectifDto/items/items')

    for mandate in mandates:
        # Create a dictionary to hold mandate data
        mandate_dict = {}

        # Extract description, dateDebut, and dateFin
        mandate_dict['descriptionMandat'] = mandate.find('descriptionMandat').text
        mandate_dict['dateDebut'] = mandate.find('dateDebut').text
        mandate_dict['dateFin'] = mandate.find('dateFin').text if mandate.find('dateFin').text is not None else 'N/A'

        # Extract all years and corresponding amounts
        remunerations = mandate.findall('remuneration/montant/montant')

        for remuneration in remunerations:
            year = remuneration.find('annee').text
            amount = remuneration.find('montant').text
            mandate_dict[f'{year}_remuneration'] = amount

        # Append the dictionary to the list
        data.append(mandate_dict)

    # Create a DataFrame
    df = pd.DataFrame(data)

    if df.empty:
      return None

    df.insert(0, 'uuid', uuid)
    df.insert(1, 'datedepot', extract_datedepot_from_xml(xml_data))

    df_long = pd.melt(
      df,
      id_vars=["uuid", "datedepot", "descriptionMandat", "dateDebut", "dateFin"],
      var_name="year",
      value_name="remuneration"
    )
    df_long['year'] = df_long['year'].str.replace('_remuneration', '')

    df_long = df_long[~df_long['remuneration'].isna()]
    df_long.reset_index(drop=True)

    # Return the DataFrame
    return df_long

```

## Extract participationDirigeantDto

```python
def parse_participationDirigeantDto(xml_string, uuid=42):
    root = ET.fromstring(xml_string)

    participation_list = []

    for items in root.findall('.//participationDirigeantDto/items'):
        for item in items.findall('items'):
            participation_info = {}

            # General Info
            participation_info['commentaire'] = item.find('commentaire').text if item.find('commentaire') is not None else None
            participation_info['nomSociete'] = item.find('nomSociete').text if item.find('nomSociete') is not None else None
            participation_info['activite'] = item.find('activite').text if item.find('activite') is not None else None
            participation_info['dateDebut'] = item.find('dateDebut').text if item.find('dateDebut') is not None else None
            participation_info['dateFin'] = item.find('dateFin').text if item.find('dateFin') is not None else None

            # Remuneration info
            remuneration = item.find('remuneration')
            participation_info['brutNet'] = remuneration.find('brutNet').text if remuneration.find('brutNet') is not None else None
            participation_info['montant'] = []

            for montant in remuneration.findall('montant'):
                for montant_detail in montant.findall('montant'):
                    montant_info = {}
                    montant_info['annee'] = montant_detail.find('annee').text if montant_detail.find('annee') is not None else None
                    montant_info['montant'] = montant_detail.find('montant').text if montant_detail.find('montant') is not None else None
                    if montant_info['annee'] is not None and montant_info['montant'] is not None:
                        participation_info['montant'].append(montant_info)

            participation_list.append(participation_info)

    df = pd.DataFrame(participation_list)
    if df.empty:
      return None
    df_exploded = df.explode('montant')

    # df_exploded['annee'] = df_exploded['montant'].apply(lambda x: x['annee'] if x is not None and 'annee' in x else None)
    # df_exploded['montant'] = df_exploded['montant'].apply(lambda x: x['montant'] if x is not None and 'montant' in x else None)

    df_exploded['annee'] = df_exploded['montant'].apply(lambda x: x['annee'] if isinstance(x, dict) and 'annee' in x else None)
    df_exploded['montant'] = df_exploded['montant'].apply(lambda x: x['montant'] if isinstance(x, dict) and 'montant' in x else None)


    df_exploded.insert(0, 'uuid', uuid)
    df_exploded.insert(1, 'datedepot', extract_datedepot_from_xml(xml_string))

    return df_exploded
```

## split into 1 declaration strings

```python
import xml.etree.ElementTree as ET

def split_xml(xml_string, chunk_size):
    root = ET.fromstring(xml_string)
    declarations = root.findall('declaration')
    num_declarations = len(declarations)
    chunks = []

    for i in range(0, num_declarations, chunk_size):
        chunk_declarations = declarations[i:i+chunk_size]
        chunk_root = ET.Element('declarations')  # Create a new root element for each chunk
        chunk_root.extend(chunk_declarations)
        chunk_string = ET.tostring(chunk_root, encoding='utf-8').decode('utf-8')
        chunks.append(chunk_string)

    return chunks


chunk_size = 1  # Number of declarations per chunk
xml_chunks = split_xml(data, chunk_size)
```

```python
from tqdm.auto import tqdm

def append_multiple_declarations(large_declaration_list):

  general_df_list = []
  participationFinanciereDto_df_list = []
  participationDirigeantDto_df_list = []
  mandatElectifDto_df_list = []

  for curr_declaration in tqdm(large_declaration_list):
    curr_uuid = extract_uuid_from_xml(curr_declaration)

    # print('done1')
    general_df_list.append(parse_general(curr_declaration, curr_uuid))
    # print('done2')
    participationFinanciereDto_df_list.append(parse_participationFinanciereDto(curr_declaration, curr_uuid))
    # print('done3')
    participationDirigeantDto_df_list.append(parse_participationDirigeantDto(curr_declaration, curr_uuid))
    # print('done4')
    mandatElectifDto_df_list.append(parse_mandatElectifDto(curr_declaration, curr_uuid))
    # print('done5')



  return pd.concat(general_df_list), pd.concat(participationFinanciereDto_df_list), pd.concat(participationDirigeantDto_df_list), pd.concat(mandatElectifDto_df_list)
```
