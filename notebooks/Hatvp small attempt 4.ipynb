```python

import requests
import xml.etree.ElementTree as ET

def parse_xml_file(url):
    # Download the XML file
    response = requests.get(url)
    xml_content = response.text
    
    # Parse the XML content
    root = ET.fromstring(xml_content)
    
    # Function to recursively parse XML elements
    def parse_xml_element(element):
        properties = {}
        
        for child in element:
            if len(child) > 0:
                # Recursively process nested elements
                nested_properties = parse_xml_element(child)
                
                # Merge nested properties into the main properties dictionary
                for key, value in nested_properties.items():
                    if key in properties:
                        properties[key].extend(value)
                    else:
                        properties[key] = value
            else:
                property_name = child.tag
                property_value = child.text
                
                if property_name in properties:
                    properties[property_name].append(property_value)
                else:
                    properties[property_name] = [property_value]
        
        return properties
    
    # Parse the XML and retrieve the properties
    parsed_properties = parse_xml_element(root)
    
    return parsed_properties

# URL of the XML file
xml_url = 'https://raw.githubusercontent.com/louispaulet/gpt4_readline/main/simple%20text/complete_declaration.txt'

# Call the function to parse the XML file
parsed_data = parse_xml_file(xml_url)

# Print the parsed properties
for property_name, item_list in parsed_data.items():
    print(f"Property: {property_name}")
    print(f"Items: {item_list}")
```
